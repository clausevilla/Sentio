```python
import pandas as pd

df = pd.read_csv("./sentio-data-processed.csv")

print(f"Total rows: {len(df)}")
print(f"Label distribution:\n{df['label'].value_counts()}")

increment_samples = []
for label in ['Stress', 'Depression', 'Suicidal', 'Normal']:
    label_df = df[df['label'] == label]
    sampled = label_df.sample(n=250, random_state=42)
    increment_samples.append(sampled)

increment_df = pd.concat(increment_samples)
train_df = df.drop(increment_df.index)

increment_df = increment_df.sample(frac=1, random_state=42).reset_index(drop=True)
train_df = train_df.reset_index(drop=True)

increment_df = increment_df[['text_preprocessed', 'label']]
train_df = train_df[['text_preprocessed', 'label']]

increment_df.to_csv("./sentio-data-increment.csv", index=False)
train_df.to_csv("./sentio-data-train.csv", index=False)

print(f"\nIncrement set: {len(increment_df)} rows")
print(f"Increment label distribution:\n{increment_df['label'].value_counts()}")
print(f"\nTrain set: {len(train_df)} rows")
print(f"Train label distribution:\n{train_df['label'].value_counts()}")
```
